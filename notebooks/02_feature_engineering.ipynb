# 02 – Feature Engineering

This notebook demonstrates the technical indicators produced by
`src/utils/indicators.py` and performs a correlation analysis.


```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.fetch_data import fetch_historical_data
from src.data.preprocess import feature_engineering_pipeline
```

## 1. Load and Engineer Features


```python
raw_df = fetch_historical_data('GP', '2022-01-01', '2024-01-01')
df = feature_engineering_pipeline(raw_df)

print('Shape after feature engineering:', df.shape)
print('\nNew columns added:', [c for c in df.columns if c not in raw_df.columns])
df.head()
```

## 2. SMA & EMA Overlay


```python
col_map = {c.lower(): c for c in df.columns}
close_col = col_map['close']

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df.index, df[close_col], label='Close', linewidth=1)
ax.plot(df.index, df['SMA_7'], label='SMA 7', linewidth=1.2)
ax.plot(df.index, df['SMA_21'], label='SMA 21', linewidth=1.2)
ax.plot(df.index, df['EMA_12'], label='EMA 12', linewidth=1.2, linestyle='--')
ax.set_title('GP – Moving Averages')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Bollinger Bands


```python
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df.index, df[close_col], label='Close', color='black', linewidth=1)
ax.plot(df.index, df['BB_Middle'], label='BB Middle', color='blue', linewidth=1)
ax.fill_between(df.index, df['BB_Lower'], df['BB_Upper'],
                alpha=0.15, color='purple', label='BB Band')
ax.set_title('GP – Bollinger Bands')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. RSI


```python
fig, ax = plt.subplots(figsize=(14, 3))
ax.plot(df.index, df['RSI'], color='purple', linewidth=1)
ax.axhline(70, linestyle='--', color='red', alpha=0.6, label='Overbought')
ax.axhline(30, linestyle='--', color='green', alpha=0.6, label='Oversold')
ax.set_title('GP – RSI (14)')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. Correlation Heatmap


```python
numeric_df = df.select_dtypes(include='number')

fig, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(
    numeric_df.corr(),
    annot=False,
    cmap='coolwarm',
    center=0,
    ax=ax,
)
ax.set_title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
